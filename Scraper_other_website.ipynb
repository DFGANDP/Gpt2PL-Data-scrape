```python
import numpy as np
import html

def html_to_utf8(text="Zachwyca pi&#281;kny &#347;wiat "):
    '''
    &#281;
    Zaczyna sie zawsze od & a konczy na ;
    nadpisac ciag usuwajac 5 znakow
    '''
    lsto = []
    lstz = []
    otwarcie = "&"
    zamkniecie = ";"
    for pos,char in enumerate(text):
        if(char == otwarcie):
            lsto.append(pos)
        if char == zamkniecie:
            lstz.append(pos)

    if len(lsto) != len(lstz):
            return text
        
    lsto = np.array(lsto)
    lstz = np.array(lstz) +1 # dodane 1 z powodu ze w tablicy [start:end] zebyu miec end trzeba podac end + 1
    #print(lsto)
    #print(lstz)
    
    for i in range(len(lsto)):
        start_point = lsto[i]
        end_point = lstz[i]
        string_to_replace = text[start_point:end_point]
        #print(string_to_replace)
        utf8_string = html.unescape(string_to_replace)
        #print(utf8_string)
        text = text.replace(string_to_replace, utf8_string)
        #print(text)
        lsto -= 5
        lstz -= 5
    
    return text

html_to_utf8()
```

```python
import numpy as np

def get_lyrics(url):
    '''
    wyciaga tekst ze strony
    '''
    def html_to_utf8(text="Zachwyca pi&#281;kny &#347;wiat "):
        '''
        &#281;
        Zaczyna sie zawsze od & a konczy na ;
        nadpisac ciag usuwajac 5 znakow
        '''
        lsto = []
        lstz = []
        otwarcie = "&"
        zamkniecie = ";"
        for pos,char in enumerate(text):
            if(char == otwarcie):
                lsto.append(pos)
            if char == zamkniecie:
                lstz.append(pos)

        if len(lsto) != len(lstz):
                return text

        lsto = np.array(lsto)
        lstz = np.array(lstz) +1 # dodane 1 z powodu ze w tablicy [start:end] zebyu miec end trzeba podac end + 1
        #print(lsto)
        #print(lstz)

        for i in range(len(lsto)):
            start_point = lsto[i]
            end_point = lstz[i]
            string_to_replace = text[start_point:end_point]
            #print(string_to_replace)
            utf8_string = html.unescape(string_to_replace)
            #print(utf8_string)
            text = text.replace(string_to_replace, utf8_string)
            #print(text)
            lsto -= 5
            lstz -= 5

        return text
            
    
    page = get(url)
    bs = BeautifulSoup(page.content, 'html.parser')
    
    lyrics = bs.find('div', class_='lyrics--text').get_text()
    lyrics = html_to_utf8(lyrics)
    lyrics = removeAccents(lyrics)
    
    
    
    
    return lyrics.strip()
lyrics = get_lyrics("https://teksciory.interia.pl/bayer-full-polskie-kwiaty-tekst-piosenki,t,528172.html")
#print(lyrics)
```

```python
import json
def get_song_artist_tittle_lyrics(url, k):
    '''
    Jako ze nie jest strona posortowana wedlugh artystow to
    pliki json beda zapisywane co strone a nie co artyste !
    '''
    j=0
    all_songs_dict = {}
    page = get(url)
    bs = BeautifulSoup(page.content, 'html.parser')
    content = bs.find('ul', class_='songs--list')
    root_site = "https://teksciory.interia.pl"
    for offer in content.find_all('li', class_='item'):
        link_to_song = offer.find('a', class_='title d-md-inline')
        song_name =  link_to_song.get_text()
        print(f"SONG NAME: {song_name}")
        artist =  offer.find('a', class_='artist -inline').get_text()
        artist = artist[3:] # for delete ' - '
        print(f"ARTIST NAME: {artist}")
        # get link to song
        try:
            if 'href' in link_to_song.attrs:
                
                url = link_to_song.get('href')
                
                caly_link = root_site+url
                
                lyrics = get_lyrics(caly_link)
                
                all_songs_dict[j] = {
                "tittle" : removeAccents(song_name),
                "artist" : removeAccents(artist),
                "year" : None,
                "album" : None,
                "lyrics" : lyrics
                }
               
                j += 1
                print(j)

        except:
            pass
    
        
    
    json_object = json.dumps(all_songs_dict, indent = 4)
    nazwa_pliku = str(artist)
    if "/" in nazwa_pliku:
        nazwa_pliku = nazwa_pliku.replace("/","")
    with open(f"{k}.json", "w") as outfile:
        outfile.write(json_object)
    # copy to gdrive

    json_object = None # zeby nie dopisywalo do pliku coraz dluzszego tylko tworzylo nowy kurwa 
    all_songs_dict = {}

        
def iteruj_teksciory():
    for i in range(1,1007):
        #print(i)
        #url = f"https://teksciory.interia.pl/szukaj?page={i}&q=teksty+polskich+piosenek&t=lyric&sort=score&dr=all"
        url = f"https://teksciory.interia.pl/szukaj?page={i}&q=tekst+polska+piosenka+utw%C3%B3r&t=lyric&sort=score&dr=all"
        k = f"plik_{i}"
        get_song_artist_tittle_lyrics(url, k)
        

iteruj_teksciory()
```
